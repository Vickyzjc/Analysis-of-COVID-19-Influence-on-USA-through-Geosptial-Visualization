```python
# code for creating choropleth map of USA states
# import plotly library
import plotly

# import plotly.express module
# this module is used to create entire figures at once
import plotly.express as px
import plotly.figure_factory as ff
import numpy as np
import pandas as pd
```

```python
df_sample = pd.read_csv('counties.csv')
```

```python
df_sample
```

```python
df_sample_ny = df_sample[df_sample['state'] == 'NY']
```

```python
df_sample_ny
```

```python
df_sample_ny.columns
```

```python
df_sample_ny['actuals.cases']
```

```python
df_sample_ny["new_vaccine_ratio"] = df_sample_ny['metrics.vaccinationsCompletedRatio'] * 100
```

```python
df_sample_ny
```

```python
values = df_sample_ny['new_vaccine_ratio'].tolist()
fips = df_sample_ny['fips'].tolist()

endpts = list(np.mgrid[min(values):max(values):4j])
colorscale = [
    'rgb(193, 193, 193)',
    'rgb(239,239,239)',
    'rgb(195, 196, 222)',
    'rgb(144,148,194)',
    'rgb(101,104,168)',
    'rgb(65, 53, 132)'
]
fig = ff.create_choropleth(
    fips=fips, values=values, scope=['New York'],
    binning_endpoints=[20, 40, 50, 60, 70], colorscale=colorscale,
    county_outline={'color': 'rgb(255,255,255)', 'width': 0.5}, round_legend_values=True,
    legend_title='Ratio', title='Vaccination Completed percentage in New York County'
)
fig.layout.template = None
fig.show()
```

```python
values = df_sample_ny['actuals.cases'].tolist()
fips = df_sample_ny['fips'].tolist()

endpts = list(np.mgrid[min(values):max(values):4j])
colorscale = [
    'rgb(193, 193, 193)',
    'rgb(239,239,239)',
    'rgb(195, 196, 222)',
    'rgb(144,148,194)',
    'rgb(101,104,168)',
    'rgb(65, 53, 132)'
]
fig = ff.create_choropleth(
    fips=fips, values=values, scope=['New York'],
    binning_endpoints=[1000, 10000, 50000, 100000, 200000], colorscale=colorscale,
    county_outline={'color': 'rgb(255,255,255)', 'width': 0.5}, round_legend_values=True,
    legend_title='Ratio', title='Vaccination Completed percentage in New York County'
)
fig.layout.template = None
fig.show()
```

```python
values = df_sample_ny['actuals.cases'].tolist()
fips = df_sample_ny['fips'].tolist()

colorscale = [
    'rgb(68.0, 1.0, 84.0)',
    'rgb(66.0, 64.0, 134.0)',
    'rgb(38.0, 130.0, 142.0)',
    'rgb(63.0, 188.0, 115.0)',
    'rgb(216.0, 226.0, 25.0)'
]

fig = ff.create_choropleth(
    fips=fips, values=values,
    scope=['New York'], county_outline={'color': 'rgb(255,255,255)', 'width': 0.5},
    legend_title='Actuals Cases in New York County'

)
fig.update_layout(
    legend_x = 0,
    annotations = {'x': -0.12, 'xanchor': 'left'}
)

fig.layout.template = None
fig.show()
```
