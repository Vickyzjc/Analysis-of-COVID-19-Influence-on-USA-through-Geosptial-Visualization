```python
# code for creating choropleth map of USA states
# import plotly library
import plotly

# import plotly.express module
# this module is used to create entire figures at once
import plotly.express as px
import plotly.figure_factory as ff
import numpy as np
import pandas as pd
```

```python
df_sample = pd.read_csv('https://raw.githubusercontent.com/datadesk/california-coronavirus-data/master/cdph-vaccination-county-totals.csv')
```

```python
df_sample
```

```python
#df_sample_1 = 
type(df_sample['fips'][0])
```

```python
list_df = []
i = 0
for name, group in df_sample.groupby(by='date'):
    if name == "2021-09-27":
        list_df.append(group)
new_df = list_df[0]
```

```python
new_df
```

```python
new_df['fips'] = new_df['fips'] + 6000
```

```python
new_df
```

```python
new_df['fully_vaccinated_percent_new'] = new_df['fully_vaccinated_percent'] * 100
```

```python
new_df
```

```python
values = new_df['fully_vaccinated_percent_new'].tolist()
fips = new_df['fips'].tolist()

'''
endpts = list(np.mgrid[min(values):max(values):4j])
colorscale = ["#030512","#1d1d3b","#323268","#3d4b94","#3e6ab0",
              "#4989bc","#60a7c7","#85c5d3","#b7e0e4","#eafcfd"]
fig = ff.create_choropleth(
    fips=fips, values=values, scope=['California'], show_state_data=True,
    colorscale=colorscale, binning_endpoints=endpts, round_legend_values=True,
    plot_bgcolor='rgb(229,229,229)',
    paper_bgcolor='rgb(229,229,229)',
    legend_title='Population by County',
    county_outline={'color': 'rgb(255,255,255)', 'width': 0.5},
    exponent_format=True,
)

'''
endpts = list(np.mgrid[min(values):max(values):4j])
colorscale = [
    'rgb(193, 193, 193)',
    'rgb(239,239,239)',
    'rgb(195, 196, 222)',
    'rgb(144,148,194)',
    'rgb(101,104,168)',
    'rgb(65, 53, 132)'
]
fig = ff.create_choropleth(
    fips=fips, values=values, scope=['California'],
    binning_endpoints=[20, 40, 50, 60, 70], colorscale=colorscale,
    county_outline={'color': 'rgb(255,255,255)', 'width': 0.5}, round_legend_values=True,
    legend_title='Fully Vaccinated Percent New by County', title='Fully Vaccinated Percent New by County'
)
fig.layout.template = None
fig.show()
```
